```python
# 1. ChatGPT 챗봇
# pip install openai

import tkinter as tk
from openai import OpenAI
import keyboard
import os
from getpass import getpass

os.environ["OPENAI_API_KEY"] = getpass("OpenAI API key 입력: ")
```

```python

# Initialize OpenAI Client
# Replace with your actual API key
client = OpenAI()  

def send_message():
    user_input = user_entry.get()
    if user_input.lower() == "exit":
        root.quit()
        root.destroy()

    else:
        messages.append({"role": "user", "content": user_input})
        completion = client.chat.completions.create(model="gpt-4o-mini", messages=messages)
        reply = completion.choices[0].message.content
        chat_log.insert(tk.END, "You: " + user_input + "\n")
        chat_log.insert(tk.END, "Assistant: " + reply + "\n")
        messages.append({"role": "assistant", "content": reply})
        user_entry.delete(0, tk.END)

# Initial system message
system_message = {
    "role": "system",
    "content": "너는 영어로만 말하는 원어민이야. 내가 영어로 대화를 걸었을 때 문장이 자연스러우면 대화를 이어나가고, 문법적으로 틀린 표현이나 어색한 부분이 있으면 바로 지적해줘."
}

messages = [system_message]

# Set up Tkinter window
root = tk.Tk()
root.title("Chat with AI")
root.geometry("1500x700")

# Create text widget to display chat log
chat_log = tk.Text(root, height=50, width=200)
chat_log.pack()

# Entry widget for user input
user_entry = tk.Entry(root, width=200)
user_entry.pack()

# Button to send message
send_button = tk.Button(root, text="Send", command=send_message)
send_button.pack()

# Register HotKey Here
# keyboard.add_hotkey('enter', send_message)

# Run the Tkinter event loop
root.mainloop()
```

```python
# 2. ElevenLabs 텍스트 음성 변환
# pip install requests

import os
import requests
from pydub import AudioSegment
from pydub.playback import play
import io

# 설정 가능한 변수
output_filename = "output_audio.mp3"

url = "yout model url"

headers = {
    "xi-api-key": "your api key",
    "Content-Type": "application/json"
}

# 문장을 입력받습니다.
text = input("텍스트를 입력하세요: ")

# 음성 생성 요청을 보냅니다.
data = {
    "text": text,
    "model_id": "eleven_multilingual_v2",
    "voice_settings": {
        "stability": 0.3,
        "similarity_boost": 1,
        "style": 1,
        "use_speaker_boost": True
    }
}

response = requests.post(url, json=data, headers=headers, stream=True)

if response.status_code == 200:
    audio_content = b""
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            audio_content += chunk

    segment = AudioSegment.from_mp3(io.BytesIO(audio_content))
    segment.export(output_filename, format="mp3")
    print(f"Success! Wrote audio to {output_filename}")

    # 오디오를 재생합니다.
    play(segment)
else:
    print(f"Failed to save file: {response.status_code}")
```
