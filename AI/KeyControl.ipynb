```python
import pyautogui
import time
import threading
import tkinter as tk
import keyboard
import random  # 랜덤한 지연을 위해 필요

# 매크로 실행 상태를 관리할 변수
macro1_running = False
macro2_running = False
macro3_running = False

# 매크로1 실행 함수 (저주/중독)
def run_macro1():
    global macro1_running
    time.sleep(random.uniform(1.0, 1.2))
    for i in range(15):
        if not macro1_running:
            break  # 중지하면 매크로 실행 중단
        pyautogui.press('6')
        time.sleep(random.uniform(0.04, 0.06))  # 0.04~0.06초 사이의 랜덤 지연
        pyautogui.press('right')
        time.sleep(random.uniform(0.04, 0.06))
        pyautogui.press('enter')
        time.sleep(random.uniform(0.04, 0.06))
        pyautogui.press('7')
        time.sleep(random.uniform(0.04, 0.06))
        pyautogui.press('enter')
        time.sleep(random.uniform(0.04, 0.06))
    macro1_running = False
    print("매크로1 종료")

# 매크로1 시작/중지 함수 (저주/중독)
def toggle_macro1():
    global macro1_running
    if not macro1_running:
        macro1_running = True
        print("매크로1 시작")
        threading.Thread(target=run_macro1).start()  # 별도의 스레드에서 매크로 실행
    else:
        macro1_running = False
        print("매크로1 중지")


# 매크로2 실행 함수 (중독)
def run_macro2():
    global macro2_running
    time.sleep(random.uniform(1.0, 1.2))
    for i in range(15):
        if not macro2_running:
            break  # 중지하면 매크로 실행 중단
        pyautogui.press('7')
        time.sleep(random.uniform(0.04, 0.06))  # 0.04~0.06초 사이의 랜덤 지연
        pyautogui.press('right')
        time.sleep(random.uniform(0.04, 0.06))
        pyautogui.press('enter')
        time.sleep(random.uniform(0.04, 0.06))
    macro2_running = False
    print("매크로2 종료")

# 매크로2 시작/중지 함수 (중독)
def toggle_macro2():
    global macro2_running
    if not macro2_running:
        macro2_running = True
        print("매크로2 시작")
        threading.Thread(target=run_macro2).start()  # 별도의 스레드에서 매크로 실행
    else:
        macro2_running = False
        print("매크로2 중지")

# 매크로3 실행 함수 (마비)
def run_macro3():
    global macro3_running
    time.sleep(random.uniform(1.0, 1.2))
    for i in range(15):
        if not macro3_running:
            break  # 중지하면 매크로 실행 중단
        pyautogui.press('5')
        time.sleep(random.uniform(0.04, 0.06))  # 0.04~0.06초 사이의 랜덤 지연
        pyautogui.press('right')
        time.sleep(random.uniform(0.04, 0.06))
        pyautogui.press('enter')
        time.sleep(random.uniform(0.04, 0.06))
    macro3_running = False
    print("매크로3 종료")

# 매크로3 시작/중지 함수 (마비)
def toggle_macro3():
    global macro3_running
    if not macro3_running:
        macro3_running = True
        print("매크로3 시작")
        threading.Thread(target=run_macro3).start()  # 별도의 스레드에서 매크로 실행
    else:
        macro3_running = False
        print("매크로3 중지")


# tkinter UI 설정
root = tk.Tk()
root.title("키보드 매크로 프로그램")
root.geometry("300x450")

# 버튼 생성
start_button = tk.Button(root, text="매크로1 시작/중지", command=toggle_macro1, width=30, height=3)
start_button.pack(pady=10)

start_button = tk.Button(root, text="매크로2 시작/중지", command=toggle_macro2, width=30, height=3)
start_button.pack(pady=10)

start_button = tk.Button(root, text="매크로3 시작/중지", command=toggle_macro3, width=30, height=3)
start_button.pack(pady=10)

# 단축키로 ` 키를 눌렀을 때 toggle_macro2 실행
# keyboard.add_hotkey('F1', toggle_macro1)
# keyboard.add_hotkey('F2', toggle_macro2)
# keyboard.add_hotkey('F3', toggle_macro3)

# tkinter 메인 루프 실행
root.mainloop()
```
