```python
def solution(k, score):
    answer = []
    total = []
    for i, daily_score in enumerate(score, start=1):
        # score   total   
        total.append(daily_score)
        total.sort(reverse=True)
        # i k  , total   append
        if i<k:
            answer.append(total[-1])
        #     k-1  answer list append
        else :
            answer.append(total[k-1])
        
    return answer

k=3
score = [10, 100, 20, 150, 1, 100, 200]
solution(k,score)
```
